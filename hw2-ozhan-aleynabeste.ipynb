# 7) Build a k-NN classifier on training data and perform models selection using 5 fold cross validation

*  Initialize k-NN classifiers with **k= 5, 10, 15**
*  Calculate the cross validation scores using cross_al_score method, number of folds is 5. 
*  Note: Xval is performed on training data! Do not use test data in any way and do not separate a hold-out validation set, rather use cross-validation.

Documentation of the cross_val_score method:

[https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.cross_val_score.html#sklearn.model_selection.cross_val_score](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.cross_val_score.html#sklearn.model_selection.cross_val_score)

*  Stores the average accuracies of these folds
*  Select the value of k using the cross validation results. 

```
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from statistics import mean

# k values
kVals = [5,10,15]

# Save the accuracies of each value of kVal in [accuracies] variable
accuracies = []

# Loop over values of k for the k-Nearest Neighbor classifier
for k in kVals:
  # Initialize a k-NN classifier with k neighbors
  knn = KNeighborsClassifier(n_neighbors=k)


  # Calculate the 5 fold cross validation scores using cross_val_score
  # cv parameter: number of folds, in our case it must be 5
  acc =cross_val_score(knn,x_train,y_train, cv = 5)
  scores = acc.mean()
  accuracies.append(scores)

  # Stores the average accuracies of the scores in accuracies variable, you can use mean method


print(accuracies)
```

# 8) Retrain using all training data and test on test set

* Predict the labels of testing data 

* Report the accuracy 

```
from sklearn.metrics import accuracy_score

# Train the best classifier using all training set
best_knn = KNeighborsClassifier(n_neighbors=15)
best_knn.fit(x_train,y_train)

# Estimate the prediction of the test data
y_predTest=best_knn.predict(x_test)

# Print accuracy of test data
print(accuracy_score(y_test, y_predTest))
```

This homeworsk aims at building a k-NN classifier using Scikit library, to classify people as bad or good for credit risks. We used German Credit Risk dataset, consisting of both numerical and ordinal/categorical features. Our problem necessites the transfering and mapping categorical/ordinal features and choosing k value in such a way that we obtain highest accuracy. When we download the data, we first define the label for both train and test set and, by dropping the label column which is Risk, from datasets we hence obtain x_train/test(features df) and y_train/test(label) After that we check if there is any missing value with isnull() function. We see that there are NaN values in housing feature hence we impute those ones for both train and set set with the mode value of housing of train data. Notice that we do not use mode of test data to avoid biased results. Before training our knn model, we must transform all categorical/ordinal features(basically non-numerical values)into numbers because kNN cannot work with them directly, unlike decision trees. We use dictionary to map the corresponding meanings of categories in the dataset for example A14 means no checking account. After mapping, We first decide which ones are the ordinal variables,it is when there is a ordering of the categories. To determine those, we need some domain knowledge.Some categories are bad for the credit such as no credit account will be the worst case hence we show it as 0 and we show other ones as 1,2 depending on the importance and 'betterness'. Better it is, higher the integer value. In this homework, Account Status, Savings Account and Employement Since, Housing and Credit History features are considered as ordinal. Note that this choice may be different, it depends on the person who writes the code but result may change accordingly. Categorical variables don't have an ordering to categories. In this homework Personal Status, Property and Other Install Plans are considered as categorical. For the  categorical data we use one hot encoder, it converts each category value into a new column and assigned a 1 or 0. This approach prevents hierarchy. Finally, we build kNN classifers with k=5,10,15 and we calculate 5 fold cross validation scores.
